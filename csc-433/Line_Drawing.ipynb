```python
import matplotlib.pyplot as plt
import numpy as np
from tabulate import tabulate
def DDA(Xo, Yo, Xn, Yn):
    deltaX = Xn - Xo
    deltaY = Yn - Yo
    slope = deltaY / deltaX
    Xp = Xo
    Yp = Yo
    Steps = max(abs(deltaX), abs(deltaY))
    Xpoints = []
    Ypoints = []
    Table = [['X','Y', 'Xr','Yr']]
    Xp1 = 0
    Yp1 = 0
    for n in range(Steps):
        if slope < 1:
            Xp1 = 1 + Xp
            Yp1 = slope + Yp
        if slope > 1:
            Xp1 = (1 / slope) + Xp
            Yp1 = 1 + Yp
        if slope == 1:
            Xp1 = 1 + Xp
            Yp1 = 1 + Yp
        Xp1r = round(Xp1)
        Yp1r = round(Yp1)
        Xpoints.append(Xp1r)
        Ypoints.append(Yp1r)
        Table.append([Xp1, Yp1, Xp1r, Yp1r])
        Xp = Xp1
        Yp = Yp1
        if  Xp1 == Xn and Yp1 == Yn :
             break
    print(tabulate(Table,headers='firstrow',tablefmt='fancy_grid'))
    plt.title("Digital Diferential Analyzer Line Algorithm")
    plt.xlabel("X-Axis")
    plt.ylabel("Y-Axis")
    plt.plot(np.array(Xpoints), np.array(Ypoints), marker=".")
    plt.show()
        
DDA(100,200,500,300)

       
```

```python
import matplotlib.pyplot as plt
import numpy as np
from tabulate import tabulate
def Bresenham(Xo, Yo,Xn, Yn):
    deltaX = Xn - Xo
    deltaY = Yn - Yo
    Pk = (2 * deltaY) - deltaX
    Steps = deltaX - 1
    Xk1,Yk1, Pk1, n = 0, 0, 0, 0
    Xpoints = []
    Table = [['Xo','Yo','Pk','Pk1','Xk1','Yk1',]]
    Ypoints = []
    while Xk1 != Xn or Yk1 != Yn:
        if Pk < 0:
            Pk1 = Pk + ( 2 * deltaY )
            Xk1 = Xo + 1
            Yk1 = Yo
        if Pk >= 0:
            Pk1 = Pk + (2 * deltaY) - (2 *  deltaX)
            Xk1 = Xo + 1
            Yk1 = Yo + 1
        Table.append([Xo, Yo, Pk, Pk1, Xk1, Yk1])
        Pk = Pk1
        Xo = Xk1
        Yo = Yk1
        Xpoints.append(Xk1)
        Ypoints.append(Yk1)
        if (n == Steps): break
        n+=1
    print(tabulate(Table,headers='firstrow',tablefmt='fancy_grid'))
    plt.title("Bresenham Line Algorithm")
    plt.xlabel("X-Axis")
    plt.ylabel("Y-Axis")
    plt.plot(np.array(Xpoints), np.array(Ypoints), marker=".")
    plt.show()
        
Bresenham(15, 6, 20, 10)
    
```
