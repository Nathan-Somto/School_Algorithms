```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
BRCData = load_breast_cancer()
df = pd.DataFrame(BRCData.data, columns=BRCData.feature_names)
df['target_class'] = BRCData.target
df.head()
```

```python
class_0_area = df.loc[df['target_class'] == 0, 'mean area']
class_1_area = df.loc[df['target_class'] == 1, 'mean area']
class_0_compact = df.loc[df['target_class'] == 0, 'mean compactness']
class_1_compact = df.loc[df['target_class'] == 1, 'mean compactness']
from scipy.stats import ttest_ind
value = ttest_ind(class_0_area, class_1_area, alternative='two-sided', equal_var=False)
t_stat_compact, p_value_compact = ttest_ind(class_0_compact, class_1_compact, alternative='two-sided', equal_var=False)
# print("T-test (two-sided) p-value:", p_value_area, t_stat_area)
print("T-test (two-sided) p-value:", p_value_compact, t_stat_compact)
value
```

```python
from sklearn.preprocessing import StandardScaler

# Split data into features and label 
X = df[BRCData.feature_names].copy()
y = df["target_class"].copy() 

# Instantiate scaler and fit on features
scaler = StandardScaler()
scaler.fit(X)
# Transform features
X_scaled = scaler.transform(X.values)
X_scaled
from sklearn.model_selection import train_test_split

# Split data into train and test
X_train_scaled, X_test_scaled, y_train, y_test = train_test_split(X_scaled,
                                                                  y,
                                                             train_size=.8,
                                                           random_state=3)
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
logistic_regression = LogisticRegression(random_state=3)
svm = SVC(random_state=3)
logistic_regression.fit(X_train_scaled, y_train)
svm.fit(X_train_scaled, y_train)

log_reg_preds = logistic_regression.predict(X_test_scaled)
svm_preds = svm.predict(X_test_scaled)
from sklearn.metrics import classification_report
model_preds = {
    "Logistic Regression": log_reg_preds,
    "Support Vector Machine": svm_preds,
}

for model, preds in model_preds.items():
    print(f"{model} Results:\n{classification_report(y_test, preds)}", sep="\n\n")
```
